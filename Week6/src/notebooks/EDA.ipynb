```python
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = data[(data[column] < lower_bound) | 
                    (data[column] > upper_bound)]
    
    return outliers

# Example for Value column
outliers_value = detect_outliers_iqr(df, "Value")
print("Number of outliers in Value:", len(outliers_value))

```

```python
col = "Value"

Q1 = df[col].quantile(0.25)
Q3 = df[col].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

plt.figure(figsize=(8,4))
sns.histplot(df[col], kde=True)

plt.axvline(lower, color='red', linestyle='--', label='Lower Bound')
plt.axvline(upper, color='green', linestyle='--', label='Upper Bound')

plt.legend()
plt.title("IQR Bounds Visualization")
plt.show()
```

```python
plt.figure(figsize=(8,2))
sns.stripplot(x=df["Value"])
plt.axvline(lower, color='red', linestyle='--')
plt.axvline(upper, color='green', linestyle='--')
plt.show()
```

```python
outlier_percent = (len(outliers_value) / len(df)) * 100
print(f"Outlier percentage in Value: {outlier_percent:.2f}%")
```

```python
col = "Value"

Q1 = df[col].quantile(0.25)
Q3 = df[col].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df[col] < lower) | (df[col] > upper)]

outlier_percent = (len(outliers) / len(df)) * 100
normal_percent = 100 - outlier_percent

plt.figure(figsize=(5,4))
plt.bar(["Normal Data", "Outliers"], 
        [normal_percent, outlier_percent])

plt.ylabel("Percentage")
plt.title("Outlier Percentage (IQR Method)")
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.pie([normal_percent, outlier_percent],
        labels=["Normal", "Outliers"],
        autopct="%1.2f%%")
plt.title("Outlier Distribution")
plt.show()
```

```python
numeric_cols = df.select_dtypes(include=['int64','float64']).columns

outlier_dict = {}

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    outliers = df[(df[col] < lower) | (df[col] > upper)]
    percent = (len(outliers) / len(df)) * 100
    
    outlier_dict[col] = percent

plt.figure(figsize=(8,4))
plt.bar(outlier_dict.keys(), outlier_dict.values())
plt.xticks(rotation=45)
plt.ylabel("Outlier Percentage")
plt.title("Outlier Comparison Across Features")
plt.show()
```
